1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# define values of x
x = np.linspace(0,10,11)
x
```

```python
# define f(x)
f = 10 * x - 5

# define values of derivative of f: f'(x)=10
f_derivative = np.full(len(x), 10)
```

```python
# plot function and its derivative
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x-5', color='violet')
plt.plot(x, f_derivative, label="f'(x) = 10", color='orange')

plt.title("Function and its derivative:")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# symbolic variable x
x_symbol = sp.symbols('x')
```

```python
# Derivative of f(x) = sin(x)
func_sin = sp.sin(x_symbol)
derivative_sin = sp.diff(func_sin, x_symbol)
```

```python
# values of argument x
x_for_cos = np.linspace(0,30,200)
```

```python
# derive numeric representation
derivative_sin_numeric = sp.lambdify(x_symbol, derivative_sin, 'numpy')
```

```python
# plot f(x) = sin(x) and its derivative
plt.figure(figsize=(10, 5))
plt.plot(x_for_cos, np.sin(x_for_cos), label="f(x) = sin(x)", color="violet")
plt.plot(x_for_cos, derivative_sin_numeric(x_for_cos), label="f'(x) = cos(x)", color="orange")

plt.title("f(x) = sin(x) and its derivative:")
plt.xlabel("x (rad)")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Derivative of f(x) = ln(x)
func_ln = sp.ln(x_symbol)
derivative_ln = sp.diff(func_ln, x_symbol)
```

```python
# values of argument x: divided into 2 sets to exclude plotting for x=0
x_for_deriv_ln1 = np.linspace(-10,-0.01,200) # f(x) = 1/x doesnt extist in f(0)
x_for_deriv_ln2 = np.linspace(0.01,10,200)
x_for_ln = np.linspace(0.01, 10, 200) # f(x) = ln(x) doesnt extist in f(0)
```

```python
# derive numeric representation
derivative_ln_numeric = sp.lambdify(x_symbol, derivative_ln, 'numpy')
```

```python
# plot f(x) = ln(x) and its derivative
plt.figure(figsize=(10, 5))
plt.plot(x_for_ln, np.log(x_for_ln), label="f(x) = ln(x)", color="violet")
plt.plot(x_for_deriv_ln1, derivative_ln_numeric(x_for_deriv_ln1), label="f'(x) = 1/x", color="orange")
plt.plot(x_for_deriv_ln2, derivative_ln_numeric(x_for_deriv_ln2), color="orange")

plt.title("f(x) = ln(x) and its derivative:")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# define values of x
x_for_quadratic = np.linspace(0, 100, 200)
```

```python
# define f(x)
func_quadratic = x_for_quadratic**2 - 5*x_for_quadratic + 1
```

```python
# derivative of f(x) = x^2-5x+1
func_quadratic_symbol = x_symbol**2 - 5*x_symbol + 1
derivative_quadratic = sp.diff(func_quadratic_symbol, x_symbol)
```

```python
# derive numeric representation
derivative_quadratic_num = sp.lambdify(x_symbol, derivative_quadratic, 'numpy')
```

```python
#
# plot f(x) = x^2 - 5x + 1 and its derivative
plt.figure(figsize=(10, 5))
plt.plot(x_for_quadratic, func_quadratic, label="f(x) = x^2 - 5x + 1", color="violet")
plt.plot(x_for_quadratic, derivative_quadratic_num(x_for_quadratic), label="f'(x) = 2x - 5", color="orange")

plt.title("f(x) = x^2 - 5x + 1 and its derivative:")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.xlim(0, 50)
plt.ylim(-6, 50)
plt.grid(True)

plt.show()
```

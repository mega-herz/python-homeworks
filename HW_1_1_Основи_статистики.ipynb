**Завдання 3.** Для того ж набору даних в змінній `sum_purchases` обчисліть **медіану** кожним з зазначених методів:  

  3.1. без використання жодних бібліотек, користуючись тільки Python.  
  3.2. використовуючи відповідний метод бібліотеки numpy.  

Виведіть результат обчислень в кожному з випадків на екран. Порівняйте результати зі значенням вибіркового середнього. Напишіть коментар, для яких задач ви б використовували вибіркове середнє, а для яких - медіану?

```python
sum_purchases_sorted = sorted(sum_purchases)
sum_purchases_sorted

sum_purchases_median = 0

if len(sum_purchases_sorted) % 2 == 0:
  # if len is even
  sum_purchases_median = (sum_purchases_sorted[len(sum_purchases_sorted) // 2 - 1] + sum_purchases_sorted[len(sum_purchases_sorted) // 2]) / 2
else:
  # if len is odd
  sum_purchases_median = sum_purchases_sorted[len(sum_purchases_sorted) // 2]

print(f'3.1. Median: {sum_purchases_median}')

print(f'3.2. Median: {np.median(sum_purchases)}')

```

На відміну від середнього, медіана належить до робастних мір статистики (robust measures), тобто, вона є менш чутливою до нетипових одиничних величин або викидів. У нашому випадку, коли у даних є одиничні нетипові величини, медіана більш придатна, щоб дати правдиве уявлення про суму типового середнього чеку, ніж середнє. У протилежних випадках, коли у даних немає нетипових величин або викидів, середнє як і медіана використовується для дослідження центральної тенденції. Порівнюючи ці дві міри, можна зробити деякі висновки про скошення (skewness) розподілу.

**Завдання 4**. Обчислість стандартне відхилення наведеної вибірки в змінній `sum_purchases` кожним з зазначених методів:  

  4.1. без використання жодних бібліотек, користуючись тільки Python.  
  4.2. використовуючи відповідний метод бібліотеки numpy.  

Виведіть результат обчислень в кожному з випадків на екран.

```python
diffs = [(el - sum_purchases_mean) ** 2 for el in sum_purchases]
sum_purchases_stdev = (sum(diffs) / (len(diffs) - 1)) ** 0.5

print(f'4.1. standard deviation: {sum_purchases_stdev}')

print(f'4.2. standard deviation: {np.std(sum_purchases, ddof=1)}')
```

**Завдання 6**. Крім даних про суму покупок, ми зібрали ще дані про кількість відвідувань кожного з кожного покупця, які знаходяться в змінній `visits`. Обчисліть коефіцієнт кореляції Пірсона для суми покупок і кількості відвідувань кожним з зазначених методів:  

  6.1. без використання жодних бібліотек, користуючись тільки Python.  
  6.2. використовуючи відповідний метод бібліотеки numpy.  
  
Виведіть результат на екран. Проаналізуйте результат.

Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
### Calculate Pearson's correlation without numpy

## 1. Calculate components for Persons Correlation

# mean for sum_purchases: sum_purchases_mean calculated above

# mean for visits
visits_mean = sum(visits) / len(visits)

# standard deviation for sum_purchases (sample): sum_purchases_stdev calculated above

# standard deviation for visits (sample)
visits_diffs = [(el - visits_mean) ** 2 for el in visits]
visits_stdev = (sum(visits_diffs) / (len(visits) - 1)) ** 0.5

# covariation between sum_purchases and visits (sample)
multiply = [(sum_purchases[i] - sum_purchases_mean) * (visits[i] - visits_mean) for i in range(len(visits))]

cov_sum_purchases_visits = sum(multiply) / (len(multiply) - 1)

## 2. Calculate Pearsons correlation
pearson_sum_purchases_visits = cov_sum_purchases_visits / (sum_purchases_stdev * visits_stdev)
print(f"6.1. Pearson's correlation between sum_purchases and visits: {pearson_sum_purchases_visits}")

### Calculate Pearson's correlation with numpy
print(f"6.2. Pearson's correlation between sum_purchases and visits: {np.corrcoef(sum_purchases, visits)[0,1]}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(sum_purchases, visits)
plt.title('Scatter Plot of Purchases vs. Visits')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

Заключення: два значення суми покупки у даних є на порядок вищими за інші та дещо ускладнюють оцінку реальної лінійної залежності між цими двома змінними.

Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
# threshold for filtering out the outliers
purchase_threshold = sum_purchases_mean + 2 * sum_purchases_stdev

# filtering out the outliers
purchases_visits_list = list(zip(sum_purchases, visits))
purchases_visits_filtered = [purchases_visits_list[i] for i in range(len(purchases_visits_list)) if purchases_visits_list[i][0] <= purchase_threshold]

# getting filtered lists
sum_purchases_filtered, visits_filtered = zip(*purchases_visits_filtered)

# calculate Pearsons's correlation
pearson_sum_purchases_visits_filtered = np.corrcoef(sum_purchases_filtered, visits_filtered)
print(f"Pearson's correlation between sum_purchases and visits of filtered data: {pearson_sum_purchases_visits_filtered[0,1]}")
```
